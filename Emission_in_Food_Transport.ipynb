```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Food_Data.csv')
```

```python
df
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df_filtered = df
df_filtered = df_filtered.rename(columns = {'area': 'country'})
df_filtered
```

```python
df_filtered = df_filtered.drop(['domain', 'domain_code', 'element_code', 'item_code', 'year_code', 'source_code', 'source', 'flag', 'note', 'flag_description'], axis=1)
```

```python
df_filtered["value"] = df_filtered["value"].round(2)
df_filtered
```

```python
list_countries = ['Angola', 'Argentina', 'Armenia', 'Australia', 'Azerbaijan',
       'Bangladesh', 'Benin', 'Bolivia (Plurinational State of)',
       'Botswana', 'Brazil', 'Burkina Faso', 'Burundi', 'Cambodia',
       'Cameroon', 'Canada', 'Chad', 'Chile', 'China', 'Colombia',
       'Costa Rica', 'Cuba', "Côte d'Ivoire",
       "Democratic People's Republic of Korea",
       'Democratic Republic of the Congo', 'Denmark', 'Ecuador', 'Egypt',
       'El Salvador', 'Eritrea', 'Eswatini', 'Ethiopia', 'Fiji',
       'Finland', 'France', 'Gabon', 'Gambia', 'Ghana', 'Guatemala',
       'Guinea', 'Guinea-Bissau', 'Guyana', 'Haiti', 'Honduras', 'India',
       'Indonesia', 'Iran (Islamic Republic of)', 'Italy', 'Jordan',
       'Kazakhstan', 'Kenya', "Lao People's Democratic Republic",
       'Lebanon', 'Lesotho', 'Liberia', 'Madagascar', 'Malawi',
       'Malaysia', 'Mali', 'Mauritania', 'Mexico', 'Morocco',
       'Mozambique', 'Myanmar', 'Namibia', 'Nepal', 'Nicaragua', 'Niger',
       'Nigeria', 'Norway', 'Oman', 'Pakistan', 'Palestine', 'Panama',
       'Paraguay', 'Peru', 'Philippines', 'Republic of Korea', 'Rwanda',
       'Saint Kitts and Nevis', 'Saint Lucia', 'Saudi Arabia', 'Senegal',
       'Sierra Leone', 'Somalia', 'South Sudan', 'Sri Lanka', 'Sudan',
       'Sweden', 'Switzerland', 'Syrian Arab Republic', 'Thailand',
       'Timor-Leste', 'Togo', 'Trinidad and Tobago', 'Tunisia', 'Turkey',
       'Uganda', 'Ukraine',
       'United Kingdom of Great Britain and Northern Ireland',
       'United Republic of Tanzania', 'United States of America',
       'Venezuela (Bolivarian Republic of)', 'Viet Nam', 'Zambia',
       'Zimbabwe']
```

```python
df_filtered = df_filtered[df_filtered['country'].isin(list_countries)]
df_filtered.reset_index(drop=True, inplace=True)
df_filtered
```

```python
df_filtered["element"].value_counts()
```

```python
element_mapping = {
    'Direct emissions (N2O)': 'Nitrous Oxide (N2O)',
    'Emissions (N2O)': 'Nitrous Oxide (N2O)',
    'Emissions (CO2eq) from N2O (AR5)': 'Carbon Dioxide (CO2)',
    'Emissions (CO2eq) (AR5)': 'Carbon Dioxide (CO2)',
    'Emissions (CH4)': 'Methane (CH4)',
    'Emissions (CO2eq) from CH4 (AR5)': 'Carbon Dioxide (CO2)',
    'Emissions (CO2)': 'Carbon Dioxide (CO2)',
    'Emissions (CO2eq) from F-gases (AR5)': 'Carbon Dioxide (CO2)'
}
df_filtered["element"] = df_filtered["element"].replace(element_mapping)
```

```python
df_filtered
```

```python
df_filtered = df_filtered.groupby(['country', 'year', 'element', 'item', 'unit']).agg({'value':'mean'}).reset_index().round(2)
```
